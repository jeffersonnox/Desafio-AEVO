```python
#@title Default title text
base = pd.read_csv('20_newsgroup.csv')
```

```python
base
```

```python


base.dropna() # Apagando linhas com valores faltantes

base = base.drop(columns=["Unnamed: 0","target", "date"]) # Retirada de colunas que não serão utilizadas
```

```python
base
```

```python
base['text']
```

```python
comment_words = ' '
```

```python
#Limpeza da coluna text com relação a caracteres especiais 

for val in base.text: 

    val = str(val) 
    val = re.sub(r"[\W|0-9]", " ",val)  
  
    
#Processo para se trabalhar de forma individual com relação a cada palavra do dataset
    tokens = val.split() 
   
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 

    for words in tokens: 
        comment_words = comment_words + words + ' '
```

```python
#Visualização das palavras que aparecem com mais frequência no dataset

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white',  
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
